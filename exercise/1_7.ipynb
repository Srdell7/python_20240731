```python
import requests
from requests import Response
factory_url = 'https://data.ntpc.gov.tw/api/datasets/c51d5111-c300-44c9-b4f1-4b28b9929ca2/json?size=1000'
try:
    res:Response = requests.request('GET',factory_url)
    res.raise_for_status()
except Exception:
    print('error')
else:
    content = res.text

from pydantic import BaseModel,RootModel,Field,field_serializer
from datetime import datetime
from pprint import pprint

class Shop(BaseModel):
    廠名:str = Field(alias='organizer')
    註冊號碼:int = Field(alias='no')
    地址:str = Field(alias='address')
    統一編號:int = Field(alias='tax_id_number')
    經度:float = Field(alias='wgs84ay',ndigit=5)
    緯度:float = Field(alias='awgs84ax',ndigit=5)
    日期:datetime = Field(alias='date')
        
    @field_serializer('日期')
    def time(self,date:datetime)->str:
        return date.strftime('%d/%m/%Y')

class Factory(RootModel):
    root:list[Shop]

shops:Factory = Factory.model_validate_json(content)
shops_list = shops.model_dump()
pprint(shops_list)

for item in shops_list:
    pprint(item,indent=3)
```
