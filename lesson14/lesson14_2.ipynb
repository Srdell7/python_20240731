```python
import numpy as np #陣列運算
import matplotlib.pyplot as plt #圖表製作
t = np.arange(0,5,0.1) #t=0到5,每0.1一單位,不包含5
y1 = np.sin(2 * np.pi * t) #數學運算sin
y2 = np.cos(2 * np.pi * t) #數學運算cos

figure = plt.figure(figsize=(6, 4)) 
axes1 = figure.add_subplot(2,1,1)
axes2 = figure.add_subplot(2,1,2)

axes1.plot(t,y1,'b-.')
axes2.plot(t,y2,'r-.')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
index = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = np.array([5, 7, 3, 4, 6]) * 1000
figure = plt.figure(figsize=(8, 5))
axes = figure.add_subplot(1,1,1)
axes.bar(index, values)
plt.show()
```
