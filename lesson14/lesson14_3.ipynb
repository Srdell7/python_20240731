```python
import random

with open('names.txt',encoding='utf-8') as file:
    names_str = file.read()
    display(names_str)
names_list = names_str.split(sep='\n')
names = random.choices(names_list,k=30)
```

```python
import numpy as np
import pandas as pd

scores = np.random.randint(60,101,size = (30,5))
list1 = pd.DataFrame(scores,
                     columns=['國','英','數','社','自'],
                     index=names)
list1
```

```python
#to_numpy()
list1.to_numpy()
```

```python
#to_csv()
list1.to_csv('grade.csv')
```
