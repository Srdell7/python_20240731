```python
import pandas as pd

list1 = pd.read_csv('grade.csv',index_col = 0)
list1
```

```python
#選取欄位
list1['國']
```

```python
#選取多個欄位
list1[['國','英','數']]
```

```python
#陣列運算
#總分
sum = list1['國'] + list1['英'] + list1['數'] + list1['社'] + list1['自']
sum
```

```python
#平均
average = sum / 5
average
```

```python
#新增欄位至原檔案
list1['總分'] = sum
list1['平均'] = average
list1
```
